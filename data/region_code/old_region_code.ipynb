````python
import requests
import pymysql
from pprint import pprint

# 기본 API 정보
url = 'https://api.odcloud.kr/api/15123287/v1/uddi:a3a4ea9d-2e34-4e32-a092-ef8feb8d71f1'
apiKeyDecoding = 'chHDedlO/C/aI5p034BuUga+/D96VJiehvTkS3N5WoxtsKrXuaCHXumBLFLUXz9k2rrMUNJrNz/DAkN/yJu/2A=='

region_list = []

sido = {
    '세종특별자치시': '세종',
    '경기도': '경기',
    '강원특별자치도': '강원',
    '충청북도': '충북',
    '충청남도': '충남',
    '전라북도': '전북',
    '전라남도': '전남',
    '전북특별자치도': '전북',
    '경상북도': '경북',
    '경상남도': '경남',
    '제주특별자치도': '제주'}

# 1페이지부터 500페이지까지 반복
for page_num in range(1, 600):
    params = {
        'serviceKey': apiKeyDecoding,
        'page': page_num,
        'perPage': 100,
        'returnType': 'json'
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        items = data['data']

    # {'법정동명': '서울특별시 종로구', '법정동코드': 1111000000, '폐지여부': '존재'}
        for item in items:
            법정동명 = item['법정동명']
            법정동코드 = item['법정동코드']
            폐지여부 = item['폐지여부']

            # '구'로 끝나고 폐지여부가 '존재'인 경우만 필터링
            if 법정동명[-1] in ('구', '군', '시') and 폐지여부 == '존재':
            # if 법정동명[-1] = '구' or '군' or '시' -> 앞에만 true(문자열)
                parts = 법정동명.split(' ')
                if len(parts) >= 2:
                    시도 = parts[0]
                    시군구 = parts[1]

                     # 서울특별시 → 서울, 부산광역시 → 부산 처리
                    if '특별시' in 시도 or '광역시' in 시도:
                        시도 = 시도[:2]

                    else:
                        시도 = sido.get(시도, 시도)


                    region_list.append({
                        '지역코드': 법정동코드,
                        '시도': 시도,
                        '시군구': 시군구
                    })

    else:
        print(f"Error on page {page_num}: {response.status_code}")



print("전체 완료 총", len(region_list), "개 지역")
pprint(region_list)
````
